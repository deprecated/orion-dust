```python
from astropy.io import fits
import numpy as np
```

```python
hdulist = fits.open("/Users/will/Work/HST-STIS/McCollum9866/o8t502010_sx2.fits")
hdulist.info()
```

```python
hdu = hdulist["SCI"]
```

```python
nx, wav0, i0, dwav = [hdu.header[k] for k in "NAXIS1", "CRVAL1", "CRPIX1", "CD1_1"]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav 
```

```python
neb_hi = hdu.data[615:625].mean(axis=0)
disk_hi = hdu.data[603:609].mean(axis=0)
disk_lo = hdu.data[592:596].mean(axis=0)
neb_lo = hdu.data[574:584].mean(axis=0)
star = hdu.data[598:603].mean(axis=0)
plot(wavs, neb_hi, label="Neb hi")
plot(wavs, neb_lo, label="Neb lo")
plot(wavs, disk_hi, label="Disk hi")
plot(wavs, disk_lo, label="Disk lo")
legend()
#xlim(5200., 10200.)
xlim(6000., 7000.)
ylim(0.0, None)
gcf().set_size_inches(15,10)
```

```python
best_neb = 0.5*(neb_lo + neb_hi)
best_disk = 0.5*(disk_hi + disk_lo)
best_neb[349:366] = best_disk[349:366]
pure_star = star - best_neb
```

```python
plot(wavs, star, label="Star")
plot(wavs, pure_star, label="Star - nebula")
legend()
gcf().set_size_inches(15,10)
```

We want the spatial profile of the star, which we get from a region of the spectrum that is free of lines.

We want to subtract off the nebular continuum, which we can model as the Ha profile multiplied by a scale factor.  Although, this needs to be modified for the part near the star/disk by interpolating across.

```python
cont_trace = np.median(hdu.data[:,300:346], axis=-1)
ha_trace = np.median(hdu.data[:,353:363], axis=-1)
nebcont_trace = ha_trace*0.017
nebcont_trace[580:620] = (nebcont_trace[580] + nebcont_trace[620])/2.
purecont_trace = cont_trace - nebcont_trace
semilogy(cont_trace)
semilogy(nebcont_trace)
semilogy(purecont_trace)
xlim(400.0, 800.0)
ylim(1e-16, 3e-13)
```

```python
jstar = purecont_trace.argmax()
purecont_trace /= purecont_trace[jstar]
```

```python
pure_star_spec = hdu.data[jstar,:] - best_neb
pure_star_2d = purecont_trace[:, None] * pure_star_spec[None, :]
```

```python
data_sub = hdu.data - pure_star_2d
```

Now we refine the estimate of the stellar spectrum using the star-subtracted nebula.

```python
disk_hi2 = data_sub[603:609].mean(axis=0)
disk_lo2 = data_sub[592:596].mean(axis=0)
plot(wavs, disk_hi2, label="Disk hi")
plot(wavs, disk_lo2, label="Disk lo")
legend()
#xlim(5200., 10200.)
xlim(6000., 7000.)
ylim(None, None)
gcf().set_size_inches(15,10)
```

```python
fits.PrimaryHDU(data_sub, hdu.header).writeto("stis-218-354-sub-star.fits", clobber=True)
```
