```python
neb_hi = hdu.data[615:625].mean(axis=0)
disk_hi = hdu.data[603:609].mean(axis=0)
disk_lo = hdu.data[592:596].mean(axis=0)
neb_lo = hdu.data[574:584].mean(axis=0)
star = hdu.data[598:603].mean(axis=0)
plot(wavs, neb_hi, label="Neb hi")
plot(wavs, neb_lo, label="Neb lo")
plot(wavs, disk_hi, label="Disk hi")
plot(wavs, disk_lo, label="Disk lo")
legend()
#xlim(5200., 10200.)
xlim(6000., 7000.)
ylim(0.0, None)
gcf().set_size_inches(15,10)
```

```python
best_neb = 0.5*(neb_lo + neb_hi)
best_disk = 0.5*(disk_hi + disk_lo)
best_neb[349:366] = best_disk[349:366]
pure_star = star - best_neb
```

```python
plot(wavs, star, label="Star")
plot(wavs, pure_star, label="Star - nebula")
legend()
xlim(6000., 7000.)
gcf().set_size_inches(15,10)
```

We want the spatial profile of the star, which we get from a region of the spectrum that is free of lines.

We want to subtract off the nebular continuum, which we can model as the Ha profile multiplied by a scale factor.  Although, this needs to be modified for the part near the star/disk by interpolating across.

```python
cont_trace = np.median(hdu.data[:,300:346], axis=-1)
ha_trace = np.median(hdu.data[:,353:363], axis=-1)
nebcont_trace = ha_trace*0.017
nebcont_trace[580:620] = (nebcont_trace[580] + nebcont_trace[620])/2.
purecont_trace = cont_trace - nebcont_trace
semilogy(cont_trace)
semilogy(nebcont_trace)
semilogy(purecont_trace)
xlim(400.0, 800.0)
ylim(1e-16, 3e-13)
```

```python
jstar = purecont_trace.argmax()
purecont_trace /= purecont_trace[jstar]
```

```python
pure_star_spec = hdu.data[jstar,:] - best_neb
pure_star_2d = purecont_trace[:, None] * pure_star_spec[None, :]
```

```python
data_sub = hdu.data - pure_star_2d
```

Now we refine the estimate of the stellar spectrum using the star-subtracted nebula.

```python
disk_hi2 = data_sub[603:609].mean(axis=0)
disk_lo2 = data_sub[592:596].mean(axis=0)
plot(wavs, disk_hi2, label="Disk hi")
plot(wavs, disk_lo2, label="Disk lo")
legend()
#xlim(5200., 10200.)
xlim(6000., 7000.)
ylim(None, None)
gcf().set_size_inches(15,10)
```

### Split into small arrays and interpolate the continuum locally

Just the strongest lines.

```python
emlines = {
           "He I 5876": 5875.64,
           "H alpha": 6562.79,
           "He I 7065": 7065.21,
           "[Ar III] 7136": 7135.78,
           "[S III] 9069": 9068.90,
           "[S III] 9531": 9530.60,
           }
           
```

```python
def sanitize(s):
    return s.replace(" ", "_").translate(None, '[]')
```

```python
nx, wav0, i0, dwav = [hdu.header[k] for k in "NAXIS1", "CRVAL1", "CRPIX1", "CD1_1"]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav 
ny, x0, j0, dx = [hdu.header[k] for k in "NAXIS2", "CRVAL2", "CRPIX2", "CD2_2"]
xpos = x0 + (np.arange(ny) - (j0 - 1))*dx 
```

```python
wav_width = 300.0
xpos_height = 5.0/3600
j1 = len(xpos[xpos < -xpos_height/2])
j2 = len(xpos[xpos <= xpos_height/2])
for emline, wav0 in emlines.items():
    stampfile = "stis-218-354-" + sanitize(emline) + ".fits"
    i1 = len(wavs[wavs < wav0 - wav_width/2])
    i2 = i1 + 62
#    i2 = len(wavs[wavs <= wav0 + wav_width/2])
    stamp = hdu.data[j1:j2, i1:i2]
    stamphdu = fits.PrimaryHDU(stamp, hdu.header)
    stamphdu.header["CRPIX1"] = hdu.header["CRPIX1"] - i1
    stamphdu.header["CRPIX2"] = hdu.header["CRPIX2"] - j1
    stamphdu.writeto(stampfile, clobber=True)
```

Now try and combine the previous two approaches.

```python
j1star, j2star = 46, 52

for emline, wav0 in emlines.items():
    stampfile = "stis-218-354-" + sanitize(emline) + ".fits"
    stamphdu, = fits.open(stampfile)
    cont_trace = 0.5*(np.median(stamphdu.data[:, 22:25], axis=1) + np.median(stamphdu.data[:, 37:40], axis=1))
    if emline == "H alpha":
        neb_estimate = 0.5*(stamphdu.data[j1star-2:j1star].mean(axis=0) + stamphdu.data[j2star:j2star+3].mean(axis=0))
        star_spec = stamphdu.data[j1star:j2star].mean(axis=0) - neb_estimate
        cont_2d = cont_trace[:, None] * star_spec[None, :] / cont_trace[j1star:j2star].mean()
        newdata = stamphdu.data - cont_2d
        neb_estimate = 0.5*(newdata[j1star-2:j1star].mean(axis=0) + newdata[j2star:j2star+3].mean(axis=0))
        star_spec = stamphdu.data[j1star:j2star].mean(axis=0) - neb_estimate
    else:
        star_spec = stamphdu.data[j1star:j2star].mean(axis=0)
    cont_2d = cont_trace[:, None] * star_spec[None, :] / cont_trace[j1star:j2star].mean()
    newdata = stamphdu.data - cont_2d
    stamphdu.data = newdata
    stamphdu.writeto(stampfile.replace(".fits", "-nc.fits"), clobber=True)
    stamphdu.data = cont_2d
    stamphdu.writeto(stampfile.replace(".fits", "-cont.fits"), clobber=True)
```

```python
hdu1 = fits.open("stis-218-354-H_alpha-nc.fits")[0]
ny, x0, j0, dx = [hdu1.header[k] for k in "NAXIS2", "CRVAL2", "CRPIX2", "CD2_2"]
xpos = x0 + (np.arange(ny) - (j0 - 1))*dx 
hdu2 = fits.open("stis-218-354-S_III_9531-nc.fits")[0]
hdu3 = fits.open("stis-218-354-S_III_9069-nc.fits")[0]
ha_trace = np.median(hdu1.data[:, 33:35], axis=1)
siii = hdu2.data + hdu3.data
siii_trace = np.median(siii[:, 33:35], axis=1)
plot(xpos*3600, siii_trace/1.5e-14, label=r"[S III] $\lambda\lambda$9069+9531")
plot(xpos*3600, ha_trace/3.2e-14, label=r"H$\alpha$ $\lambda$6563")
fill_betweenx([0,1.5], [-0.1, -0.1], [0.1, 0.1], color="k", alpha=0.2)
ylim(0.0, None)
xlim(-2.5, 2.5)
legend(loc="lower right")
xlabel("Offset, arcsec")
ylabel("Relative brightness")
grid()
gcf().set_size_inches(12,4)
gcf().tight_layout()
gcf().savefig("stis-218-354-compare.pdf")
```

So the above shows that the Ha and [S III] brightness profiles are indistinguishable.   This could be interpreted as evidence for gray extinction.  On the other hand, it could just be because the edges are unresolved.
